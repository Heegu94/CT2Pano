```python
import os, glob
import numpy as np
import matplotlib.pyplot as plt

from scipy.ndimage.interpolation import rotate
```

```python
ct = np.load('C://Users//Heeguen//Desktop//heeguen//Dental_panorama_2020//np_data/ct.npy', allow_pickle=True)
```

```python
for i in range(len(ct)):
    print(i, '//', ct[i].min())
```

```python
tot_sum = np.zeros((300, 670, 670))

# for i in range(len(ct)):
for i in range(10):
    trans_ct = np.transpose(ct[i], (2,0,1)) 
    inter_ct = trans_ct[:300, :670, :670] - trans_ct.min()
    tot_sum += inter_ct
    print('min = ', tot_sum.min(), '/// max = ', tot_sum.max())
    print('................................')
    
```

```python
# from __future__ import print_function
from ipywidgets import interact
import ipywidgets as widgets
def show(x):
    idx = x
    plt.figure(figsize=(12,12))
    plt.imshow(np.rot90(tot_sum[idx, ...]), cmap='gray')

interact(show, x=widgets.IntSlider(min=0, max=(tot_sum.shape[0]-1), step=1,value=0))
```

```python
view = np.transpose(tot_sum, (2,0,1))

proj = np.zeros(view.shape[1:])
for s in range(view.shape[0]):
    proj += view[s, ...]

plt.figure(figsize=(12,6))
plt.imshow(np.flipud(proj), cmap='gray')
```
